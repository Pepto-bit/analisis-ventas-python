```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer archivo CSV
df = pd.read_csv("ventas.csv")

# Crear columna de ingresos
df["ingresos"] = df["cantidad"] * df["precio"]

# Ingresos totales
ingresos_totales = df["ingresos"].sum()

# Producto más vendido
ventas_por_producto = df.groupby("producto")["cantidad"].sum()
producto_top = ventas_por_producto.idxmax()

print("Ingresos totales:", ingresos_totales)
print("Producto más vendido:", producto_top)

# Gráfica
ventas_por_producto.plot(kind="bar")
plt.title("Ventas por Producto")
plt.xlabel("Producto")
plt.ylabel("Cantidad vendida")

plt.show()
plt.savefig("grafico_ventas.png")
```
